# BALANCE-NM Run Inspection

This notebook reads persisted CLI artifacts only; it does not execute adaptive acquisition.

```python
from pathlib import Path
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt

run_dir = Path('../results/run_000')
truth = xr.open_zarr(run_dir / 'hidden_truth.zarr')
prediction = xr.open_zarr(run_dir / 'prediction.zarr')
metrics = pd.read_csv(run_dir / 'metrics.csv')
metrics.tail()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
truth.true_interface.plot(ax=axes[0]); axes[0].set_title('True interface')
prediction.phase_probability.plot(ax=axes[1]); axes[1].set_title('Phase probability')
prediction.interface_weight.plot(ax=axes[2]); axes[2].set_title('Interface weight')
plt.tight_layout()
```
